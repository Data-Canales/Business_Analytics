##### Vamos a calcular la distancia a la puerta del sol como pto de interés. Latitud = 40.4 Longitud = -3.7

### Distancia de Haversine

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radio de la Tierra en kilómetros
    
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    
    a = sin(dlat / 2)**2 + cos(lat1)*cos(lat2)*sin(dlon / 2)**2
    c = 2 *asin(sqrt(a))
    
    return  R * c

```

```python
lat1 = 40.4167278
lon1 = -3.7033387

df['sol'] = df.apply(lambda registro: haversine(lat1,lon1,registro.latitude,registro.longitude), axis=1)
```

```python
df.groupby('distrito').sol.mean().sort_values()
```

#### vemos que la distancia mas cercana a la puerta del sol es el centro, y la más lejana es barajas a 11.5 km
